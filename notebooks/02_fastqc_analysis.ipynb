# Analisi della qualità con FastQC

Il controllo qualità delle reads è un passo essenziale prima di procedere con l'analisi. Utilizzeremo **FastQC** per generare un report sulla qualità delle reads contenute nel file `sample.fastq`.

FastQC produce due output: un file `.html` con un report interattivo e un file `.zip` con i dati grezzi. Poiché Binder potrebbe non consentire la visualizzazione diretta dell'HTML, ci concentreremo sui file di log generati da FastQC.

Per eseguire FastQC su un file FASTQ, digita `fastqc nome_file.fastq`. Puoi specificare l'opzione `--outdir` per indicare dove salvare i risultati.


```
# Esegui FastQC sul file di esempio
!mkdir -p results/fastqc
!fastqc data/sample.fastq --outdir=results/fastqc

# Elenca i file di output
a_file = !ls results/fastqc
print('Output FastQC:', a_file)
```

Dopo aver eseguito FastQC, dovresti trovare un file `.zip` e un file `.html` nella cartella `results/fastqc`. L'HTML può essere scaricato e aperto nel browser per visualizzare grafici come la distribuzione della qualità per posizione, la frequenza di GC e l'andamento della sequenza duplicata.
