# Introduzione alla linea di comando

In questa lezione imparerai i comandi di base della **shell Bash**. L'obiettivo è saper navigare nel filesystem, esaminare file di testo e combinare comandi tramite pipe (`|`). Questi concetti sono fondamentali per la bioinformatica perché quasi tutti i software per sequenziamento lavorano da linea di comando.

## Comandi fondamentali

- `pwd` — stampa la directory corrente.
- `ls` — elenca i file nella directory.
- `cd` — cambia directory.
- `head` / `tail` — mostra le prime o ultime righe di un file.
- `wc -l` — conta le righe di un file (utile per contare le reads FASTQ: ogni read occupa 4 righe).
- `grep` — filtra righe che corrispondono a un pattern.
- `cut`, `sort`, `uniq` — manipolano colonne, ordinano e contano elementi.

### Esempio
Qui sotto trovi un esempio di comandi Bash eseguiti all'interno di un notebook. Per eseguire un comando di shell in un notebook Python, anteponi `!` al comando. I risultati verranno mostrati sotto la cella di codice.


```
!pwd
!ls -l data
!head -n 4 data/sample.fastq
```
